```python
n = list(range(0,50))
five_division = []
for i in n:
    if n[i]%10==0 or n[i]%5==0:
        five_division.append(n[i])
five_division
```

```python
#8. Read the list using the enumerator.
```

```python
new_list=(["bahar","sevda","melek"])
```

```python
names = 0
for index in range(len(new_list)):
    print(names,new_list[index])
    names+=1
```

```python
for key,value in enumerate(new_list):
    print(key,value)
```
